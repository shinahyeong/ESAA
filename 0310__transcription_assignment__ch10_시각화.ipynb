# **시각화**

# [개념 정리]
**1. 시각화를 시작하며 - 맷플롯립과 시본 개요**

**맷플롯립(Matplotlib)**
- 파이썬의 대표적인 시각화 라이브러리
- MATLAB 스타일의 인터페이스가 반영된 라이브러리로 pyplot 모듈을 사용
- 단점 : 직관적이지 못한 개발 API로 인해 시각화 코딩에 익숙해지는 데 많은 시간이 필요하고, 차트의 축 이름, 차트 타이틀, 범례(legend) 등의 부가적인 속성까지 코딩 해주어야 함

**시본(Seaborn)**
- 맷플롯립보다 쉬운 구현, 수려한 시각화, 편리한 판다스(Pandas)와의 연동이 가능한 특징을 가짐
- 맷플롯립을 기반으로 하나 맷플롯립보다 상대적으로 적은 양의 코딩으로도 수려한 시각화 플롯을 제공함
- 판다스의 칼럼명을 기반으로 자동 축 이름 설정 등 편리한 연동 기능을 가짐
- 단점 : 맷플롯립 기반으로 특정 요소들의 경우 맷플롯립 함수들을 그대로 사용하여 맷플롯립에 대하여 어느정도 지식이 필요


**2. 맷플롯립(Matplotlib)**

**맷플롯립의 pyplot 모듈의 이해**
- 파이썬 시각화 기반 모듈인 pyplot을 제공

  : **import matplotlib.pyplot as plt**로 임포트
- 직선 그래프 그리기 : plot(x좌표 , y좌표)
- 차트 타이틀 설정 : plt.title('차트 이름')
- 선 그래프 출력 : plt.show()

**pyplot의 두 가지 중요 요소 - Figure와 Axes 이해**

1)** Figure 객체** : 그림을 그리기 위한 캔버스의 역할
- 그림판의 크기를 조절하거나 플롯을 최종적으로 화면에 나타내는 역할을 수행함
- 실제 그림을 그리는 작업을 수행하지 않음

2) Axes 객체 : 실제 그림을 그리는 메서드를 가짐
- 대부분의 시각화를 수행하는 메서드들이 호출됨

  (대부분의 함수 호출이 여기로 진행됨)
- X축, Y축, 타이틀, 범례 등의 속성을 설정하는데 이용
- 대다수 시각화 작업이 여기서 진행

**Figure와 Axis의 활용**

1) Figure
- **plt.figure()**: 전체 그림판의 크기를 조절하는 데 사용됨 [**figure** 객체 반환]
  - **figsize = (가로크기, 세로크기)** : 해당 크기를 가지는 Figure 객체 반환
  - **facecolor = '배경색'** : 그림판의 배경 색깔을 바꿈
  
  → 단, 전체 그림판의 색이 변하는 것이 아닌, Axes가 차지하는 영역을 제외한 영역에 나타남

2) Axes
- **plt.axes()** : 현재 사용하는 Axes 객체를 반환함 [**AxesSubplot**라는 Axes의 실체화된 하위클래스가 반환]
  - 따로 plt.show()를 하지 않아도 그림이 표출됨

3) Figure와 Axes 활용

(1) Figure와 Axes 동시 반환
- **fig, ax = plt.subplots()**

(2) Axes 객체 메서드를 이용한 시각화
- **ax.plot(x축, y축)** : 선 그래프를 그림
- **ax.set_title('제목')** : 선 그래프의 타이틀 설정

→ 해당 2 가지 방법을 이용하여 앞선 plt.plot()과 plt.title()을 이용하여 그린 선 그래프와 동일한 결과를 얻을 수 있음

**여러 개의 plot을 가지는 subplot들을 생성하기**
**plt.subplots() **
- 주로 여러 개의 subplot을 생성하는데 활용됨
- 여러개의 그래프를 그릴 수 있게 하는 1 개의 Figure와 여러 개의 Axes를 생성하고 반환함
- 주요 인자 : nrows, ncols, figsize
  - **nrows** : 전체 subplot들의 배치를 2차원 행렬 형태로 표현할 때 행의 개수
  - **ncols**: 열의 개수
  - **figsize** : 모든 subplot들을 포함한 전체 Figure 크기를 설정함
- Axes 객체의 가능한 형태 : 튜플, 넘파이의 ndarray

**pyplot의 plot( )함수를 이용하여 선 그래프 그리기**
**plt.plot()**
- 선그래프를 그릴 때 활용
- plot() 함수의 X 좌표, Y 좌표
  - 적용 가능 값 : 파이썬 리스트, 넘파이 ndarray, 판다스(Pandas)의 DataFrame/ Series
  - X 좌표 값과 Y 좌표 값의 개수는 모두 같은 크기를 가져야 함
    
    → if not, 오류 반환!
- color 인자를 이용한 선 색깔 변경 가능
  - API 기반의 시각화 구현이기 때문에 함수의 인자들에 대하여 알고 있어야 설정 가능  

<결론>

plot() 함수의 시각화

→ **plot(x_value, y_value, color='색깔', marker='좌표점 표현', linestyle='선 형태', linewidth='선 두께', markersize = '좌표점 크기')** 과 같이 표현

**축 명칭 설정, 축의 눈금(틱)값 회전, 범례(legend) 설정하기**

[ plot()을 이용한 시각화 ]
- matplotlib의 경우 기본적으로 축의 명칭을 자동으로 설정하지 않음
  - 축의 명칭을 설정하고 싶다면 **pyplot의 xlabel(), ylabel()**과 같은 함수를 이용해야 함
- 축의 눈금값(틱값)의 회전
  - 회전시키는 이유 : 축의 값이 문자열이고 눈금값이 많은 경우 눈금 간의 값이 중복되면서 제대로 보이지 않음
  - **xticks(), yticks()** 함수 이용 (대개 x축 눈금값만 회전)
- 범례(legend) 설정
  - 여러 개의 그래프/차트를 한 Axes 내에서 표시할 때 이들을 구분할 때 편리하게 이용됨
  - 맷플롯립의 경우 여러 개의 플롯을 하나의 Axes 내에서 그릴 수 있음
  - **plot() 내 label의 객체**로 나타냄

  → 이후 **plt.legend()**를 통하여 label로 설정된 문자열을 범례로 표시함

[ Axes 객체 메서드를 이용한 시각화 ]
- plt.axes() : Axes 객체 추출
- plt → ax로 변경하여 각각의 plot과 bar 차트를 시각화 함
- title을 추출하는 경우 ax.title이 아닌 ax.set_title()로 추출
- label도 마찬가지로 set_xlabel(), set_ylabel()로 추출
- 하지만 legend의 경우 set_legend가 아닌 legend()로 추출해야 함

**여러 개의 subplots들을 이용하여 개별 그래프들을 subplot별로 시각화하기**
- 많은 feature로 이뤄진 데이터 세트를 분석하는 경우, 서로 다른 feature 간의 그래프 비교, 같은 feature에서 서로 다른 그래프를 이용한 여러가지 특성 분석에 효과적
- 여러 개의 subplots들을 이용한 시각화 구현 시 유의사항
  - **개별 Axes 객체를 직접 이용**하여 시각화를 구현해야 함
  
  → 그래프를 그리는 것에 국한되는 것이 아닌 타이틀, 축의 명칭 눈금값, 범례 설정 시에도 개별 Axes에 직접 이를 적용해야 함
  - 각각의 subplots()에는 **배열 인덱싱 방식**으로 접근해야 함

  → nrows가 1이거나 necols가 1이면 1차원 배열 인덱싱 방식, 2 이상이 되면 2차원 배열 인덱싱 방식으로 Axes 객체에 접근해야 함

# [코드 필사]

```python
# pyplot 모듈을 (관용적으로) plt라는 이름으로 import.
import matplotlib.pyplot as plt

# x축 입력값 [1, 2, 3], y축 입력값 [2, 4, 6]으로 선 그래프 생성
plt.plot([1, 2, 3], [2, 4, 6])

# 선 그래프 타이틀을 Hello plot으로 설정.
plt.title("Hello plot")

# 선 그래프를 출력
plt.show()
```

```python
# plt.figure()는 주로 figure의 크기를 조절하는 데 사용됨
# 아래는 figure 크기가 가로 10, 세로 4인 Figure 객체를 설정하고 반환함.
plt.figure(figsize=(10, 4))

plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

```python
plt.figure(figsize=(4, 4), facecolor='yellow')
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

```python
ax = plt.axes()
print(type(ax))
```

```python
### pyplot에서 설정된 Figure와 Axes 객체를 함께 가져오기
fig, ax = plt.subplots()
print('fig type:', type(fig), '\nax type:', type(ax))
```

```python
fig, ax = plt.subplots(figsize=(4, 4))

# Axes 객체의 plot()을 이용하여 선 그래프를 그림.
ax.plot([1, 2, 3], [2, 4, 6])

# Axes 객체의 set_title()을 이용하여 선 그래프 타이틀을 Hello plot으로 설정
ax.set_title('Hello plot')
plt.show()
```

```python
# x 좌표 값을 0~99까지 100개로 설정.
x_value = np.arange(0, 100)
# y 좌표 값은 개별 x 좌표 값을 2배로 해서 100개로 설정.
y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축 눈금값을 90도로 회전
plt.xticks(rotation=90)
plt.show()
```

```python
x_value = np.arange(0, 100)
y_value = 2*x_value
plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축의 눈금값을 np.arange(0, 100, 5)로 하여 0부터 99까지 5 스텝으로 20개를 표시하고 90도 회전
plt.xticks(ticks = np.arange(0, 100, 5), rotation=90)

plt.title('Hello plot')
plt.show()
```

```python
# X좌표값은 0~99까지 100개, 첫 번째 Y좌표값은 X좌표값의 두 배, 두 번째 좌표값은
# X좌표값의 4배
x_value_01 = np.arange(0, 100)
y_value_01 = 2*x_value_01
y_value_02 = 4*x_value_01

# X좌표값과 첫 번째 Y좌표값으로 녹색의 선 그래프를 그림. label을 'green line'으로 설정.
plt.plot(x_value_01, y_value_01, color='green', label='green0 line')
# X좌표값과 두 번째 Y좌표값으로 빨간색 선그래프를 그림. label을 'red line'으로 설정.
plt.plot(x_value_01, y_value_02, color='red', label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')
# 개별 plot들의 label로 설정된 문자열을 범례로 표시
plt.legend()
```

```python
x_value_01 = np.arange(1, 10)
y_value_01 = 2*x_value_01

# 마커를 포함한 빨간색 대시 선 그래프를 그리고, label은 'red line'으로 설정
plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
# X값에 따른 Y값을 나타내는 초록색 막대 그래프를 그리고, label은 'bar plot'으로 설정
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 두 개의 그래프에서 label 설정된 문자열 값을 범례로 표시
plt.legend()

plt.title('Hello plot')

plt.show()
```

```python
figure = plt.figure()
# Axes 객체를 추출하여 ax 변수에 할당
ax = plt.axes()

# plt.plot()은 ax.plot()으로 plt.bar()는 ax.bar()로 변경.
ax.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')

# plt.xlabel()은 ax.set_xlabel()로, plt.ylabel()은 ax.set_ylabel()로 변경
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# ax.set_legend()가 아니라 ax.legend()임
ax.legend()
# plt.title()을 plt.set_title()변경
ax.set_title('Hello plot')
plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2*x_value_01
y_value_02 = 2*x_value_02

# figsize는 (6, 3)인 Figure와 2개의 Axes를 가지는 subplots을 반환함.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

# 개별 Axes 객체들을 각각 이용하여 선 그래프와 막대 그래프를 그림.
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')
# 개별 Axes 객체 각각에 x축명을 설정.
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')
# 개별 Axes 객체 각각에 범례를 설정
ax[0].legend()
ax[1].legend()
# 개별 Axes 객체 각각에 타이틀을 설정
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_02 = 2*x_value_01
y_value_02 = 2*x_value_02

# figsixe는 (6, 6)인 Figure와 2*2개의 Axes를 가지는 subplots을 반환.
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

# 2*2 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02,y_value_02, color='red', label='red bar')

# 개별 Axes 객체 각각에 X축명을 설정.
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] xaxis')
ax[1][1].set_xlabel('ax[1][1] x axis')

# 개별 Axes 객체 각각에 범례를 설정
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
