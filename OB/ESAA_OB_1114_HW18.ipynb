#5장. 합성공 신경망1

##5.1 합성곱 신경망
- 딥러닝의 역전파: 순전파 과정에 따라 계산된 오차 정보가 신경망의 모든 노드로 전송됨
  - 이런 계산 과정은 복잡하고 많은 자원을 요구함
  - 계산하는 시간도 오래걸림
    
    -> 이를 해결하는 방법이 합성곱 신경망임
    
###5.1.1 합성곱층의 필요성
- 합성곱 신경망: 이미지 전체를 한 번에 계산하는 것이 아닌 이미지의 국소적인 부분을 계산함으로써 시간과 자원을절약하여 세밀한 부분까지 분석할 수 있는 신경망
  - 이미지나 영상을 처리하는데 유용함
    - 이미지 분석: 펼쳐서 각 픽셀에 가중치를 곱하여 은닉층으로 전달하게 됨
      - 이렇게 펼쳐서 분석하면 데이터의 공간적 구조를 무시하는 상황이 발생함 -> 이를 방지하고자 합성곱층을 도입

###5.1.2 합성곱 신경망의 구조
- 합성곱 신경망: 주로 음성인식, 이미지/영상 인식에서 주로 사용되는 신경망
  - 다차원 배열 데이터를 처리하도록 구성되어 컬러 이미지 같은 다차원 배열 처리에 특화
  - 계층 다섯개로 구성: 입력층, 합성곱층, 풀링층, 완전 연결층, 출력층
- 합성곱 신경망은 합성곱층과 풀링층을 거치며 입력 이미지의 주요 특성 벡터를 추출함
  - 이후 추출된 주요 특성 벡터들은 완전 연결층을 거치며 1차원 벡터로 변환되고, 마지막에 출력층에서 활성화 함수인 소프트맥스 함수를 사용해 결과를 출력함

**입력층**
- 입력 이미지 데이터가 최초로 거치게 되는 계층
  - 이미지는 단순 1차원이 아닌 높이, 너미, 채널의 값을 갖는 3차원 데이터
    - 이때 채널은 이미지가 그레이스케일이면 1값을 가지고, 컬러이면 3값을 가짐

**합성곱층**
- 입력 데이터에서 특성을 추출하는 역할을 수행함
  - 특성 추출: 입력 이미지가 들어왔을 때 이미지에 대한 특성을 감지하기 위해 커널이나 필터를 이용
    - 커널/필터: 이미지의 모든 영역을 훑으며 특성을 추출하게 됨 -> 이렇게 추출된 결과물을 '특성 맵'이라 부름
    - 커널은 스트라이드라는 지정된 간격에 따라 순차적으로 이동하며 적용됨

**풀링층**
- 합성곱층과 유사하게 특성 맵의 차원을 다운 샘플링하여 연산량을 감소시키고, 주요한 특성 벡터를 추출하여 학습을 효과적으로 할수 있게 함
- 풀링 연산
  - 최대 풀링: 대상 영역에서 최댓값을 추출
  - 평균 풀링: 대상 영역에서 평균을 바환
    - 대부분 합성곱 신경망에선 최대 풀링을 이용


**완전연결층**
- 합성곱층과 풀링층을 거치며 차원이 축소된 특성 맵은 최종적으로 완전 연결층으로 전달됨
  - 이때 이미지는 3차원 벡터에서 1차원 벡터로 펼쳐지게 됨

**출력층**
- 소프트맥스 활성화 함수가 사용되는 부분으로, 입력받은 값을 0~1사이의 값으로 출력함
  - 마지막 출력층의 소프트맥스 함수를 사용해 이미지가 각 레이블에 속할 확률값이 출력됨
  - 가장 높은 확률 값을 갖는 레이블이 최종값으로 선정됨


###5.1.3 1D, 2D, 3D 합성곱
- 합성곱은 이동하는 방향의 수와 출력 형태에 따라 1D, 2D, 3D로 나뉨

**1D 합성곱**
- 필터가 시간을 축으로 좌우만 이동할 수 있는 합성곱
  - 입력과 필터에 대한 출력은 W(입력)이 됨
  - 그래프의 곡선을 완화할 때 많이 이용

**2D 합성곱**
- 필터가 방향 두 개로 움직이는 형태를 나타냄
  - 입력과 필터에 대한 출력은 입력(W, H)가 됨
  - 출력 형태가 2D 행렬이 됨

**3D 합성곱**
- 필터가 움직이는 방향이 3개로 나뉨
  - 입력에 대해 필터를 적용하면 출력으로 입력(W,H,L)을 갖는 형태
  - 이때 d<L을 유지하는 것이 중요!!

**3D 입력을 갖는 2D 합성곱**
- 입력이 3D 형태 임에도 출력 형태가 3D가 아닌 2D행렬을 취하는 형태
  - 필터에 대한 길이가 입력 채널의 길이와 같으면 이런 형태가 생성
  - 입력에 필터를 적용하면 출력은 입력으 앞 2자리(W, H)가 됨
  - 필터는 마찬가지로 두 방향으로면 움직이며, 출력형태는 2D가 됨
  - 대표 사례: LeNet-5, VGG

**1*1 합성곱**
- 3D 형태로 입력됨
  - 입력에 필터를 적용하면 출력은 입력의 앞 2자리(W, H)가 됨
- 이러한 합성곱 형태에선 채널 수를 조정해 연산량이 감소되는 효과가 존재
- 대표 사례: GoogleNet
