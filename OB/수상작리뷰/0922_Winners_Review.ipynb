## 서울 특별시 열섬현상 연구 - 행정동별 원인분석 및 맞춤형 완화 방안

- 링크: https://data.kostat.go.kr/sbchome/bbs/boardList.do?boardId=SBCSBBS_000000017002&curMenuNo=OPT_08_06_00_0
- 연구배경 및 목적
  - 주제 선정 배경:
    - 기후변화 가속화
    - 도시 열섬 현상

    -> 두가지 다 논문을 참고하거나 기상자료 개방 포털 등 근거 있는 자료를 사용함
  - 연구 범위
    - 서울 특별시
    - 행정동 단위

    -> 각각의 단위를 왜 이렇게 구성하였는지 이유를 간략하게 나타냄 + 범위를 정해놓으므로써 연구할 때 해당 범위 내에 집중 할 수 있음
  - 분석 전략: 행정동 단위의 세밀한 열분포와 원인을 진단하여 지역별 맞춤 열섬 완화 정책 수립에 기여하는 종합적 접근법 개발

- 데이터: AWS관측값과 행정동 경계 데이터 등 다양한 원천 자료를 융합하여 종합적 열분포를 재구성함
  - 다양한 데이터를 선정했는데 여기에 데이터 선정기준을 명확히 밝힘으로써 신뢰성을 확보함 + 어떤 변수를 종속변수로 도출할 것인지, 어떤 변수를 독립변수로 사용할 것인지 명확히 명시함
- 데이터 전처리
  - 데이터 결합
  - 행정동 기온 보간(크리깅 기법 이용)
  - UHII 계산 : 열섬강도지수 정의및 산출 수행
  - 독립변수 결합 -> 카카오 API를 이용하여 측정소의 주소를 좌표로 변환함  
  - 최종 데이터셋 생성
- 모델링
  - UHII 분포 시각화를 통해 고위험 지역을 도축
    - 여의도 인근, 아차산 인근, 성남시 인접 지역 등이 도출됨
  - 회귀 모델링을 통한 위험지역 및 원인 식별
    - 다양한 모델들을 생각을 했고, 이들을 직접적으로 성능을 비교하여 최고 성능으로 선정 (모형별 성능을 시각화로 보여줌)
  - 정책 수립에 즉시 적용 가능한 기초자료를 제공함
- 분석 결과
  - 분석 결과 부분이 약간은 짝ㅂ음 -> 2장정도( 나머지는 5-6장정도)
  - 분석 결과를 중요하게 생각하는 심사위원이 많기에, 이부분을 늘리면 좋았겠다고 생각이 듦
- 정책활용전략은 체계적으로 잘작성되어 있어 좋았으나, 글씨가 많고, 핵심을 찾기 어려워보였음 -> 핵심적인 부분을 직접 도출하여 이를 찾아내는 것도 중요하다고 생각함
