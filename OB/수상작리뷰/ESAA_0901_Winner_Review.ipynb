##2025년 통계데이터 활용대회 대상_보고서 수상작 리뷰

링크: https://data.kostat.go.kr/sbchome/bbs/boardDetail.do

0. 보고서 흐름

연구요약 - 주제선정배경 및 연구의 필요성 - 연구 목적(전략) - 활용 데이터 -  3가지 분석과정+분석 결과 나열 - 활용 타당성 검증(분석 활용) -  기대 효과 및 방향성 제시 - 서비스 예시(실제로 상용화버전 프로토 타입)  

1. 주제 선정 및 연구 목적
- 주제: 소상공인 생존율 높이는 데이터 전략 - 소비자 성향 기반 상권 분석 프레임워크
- 최근 이슈화 되는 인구사회학적 문제를 바탕으로 선정
- 소상공인의 구조적 위기 심화 + 기존 상권 분석 시스템의 보완 필요성이라는 '사회문제 + 기존 시스템의 문제' 전형적인 내용을 통해 주제 선정 배경 및 연구의 필요성 부분을 생각함

2. 활용 데이터
- 데이터 / 출처 / 주요 내용 / 활용 목적의 표 형태로 제시
- 데이터의 연도를 대체적으로 균일하게 맞춤 (지표의 신빙성 확보)
- 데이터 별 활용 목적을 명확히 밝혀 어떤 분석에 어떤 데이터를 활용하였는지 투명하게 확인 가능

3. 분석 과정

[1] 상권 분석

(1) 상권 클러스터링 분석
- 3단계로 진행하였다는 걸 깔끔하게 보여줌 (파생변수 생성 / K-평균 군집분석 / ELBOW 기법)
- 대체적으로 다양한 최신 논문을 활용하여 해당 분석을 왜 수행하였는지 타당성을 확보함 (예: K-평균 군집분석은 각 군집의 특성을 직관적으로 해석할 수  있기 때문에 이를 이용해 군집별로 중심점을 찾고자 활용함)
- 간단하게 분석 과정을 나타낸 후 서울/경기도 각각의 분석 결과를 표 및 지도 형태로 제시 (업종과 고객 비중을 각 순위별로 나타내어 상권을 분류함)

(2) 업종별 소비 성향 분석 과정
- 소비 성향 점수를 산출 (이 점수를 어떤식으로 나타내게 되었는지는 안나와 있고, 이를 사용하게 된 목적과 이를 어떤식으로 해석하는지, 결과가 어떤식으로 나타나는지 간단한 그래프와 해석을 나타냄)
- 이를 통해 업종군별/ 성,연령대별 소비층과 선호도를 분석함

(3) 업종별 종합점수 분석
- 평균 소비금액, 건당평균금액, 평균 이용건수, 소비 강도를 통해 이에 대한 종합점수를 산출함 (Min-Max정규화를 활용하고 이를 동일 가중치를 부여한 종합점수 산출)
- Z-score 활용

[2] 공급 분석

(1) 입지 계수와 매출 격차를 활용하여 두 짗표를 종합적으로 고려함 (해당 지표는 논문을 참고한 것으로 나타냄) - 계산하기 편리하게 생겼음
- 어떤 지역의 수요와 공급의 격차가 발생하는지를 나타내여 상권에서 유망한 업종을 추천함
- 이걸 생존율 분석으로 활용

[3] 수요 분석

(1) 관심사 점수 산출
- 아파트 단지별 소비 기반으로 소비 관심군을 파악하고 해당 지역의 수요층을 파악함


전체적인 평
- 피피티는 최대한 깔끔하고, 핵심만 담되, 내용이 빠지면 안됨
- 처음에 연구 요약부분이 굉장히 임팩트 있음 -> 상권 분류를 통해서 해당 상권을 분석하고, 이에 대한 수요와 공급을 각각 확인ㄴ하여 입지에 대한 업종을 추천하는 전략으로 간다는 걸 시각화를 통해 전체적 파이프라인을 확인시켜줌

-> 아직 시작 하기 전에 미리 어떤 식으로 발표를 진행한다가 눈에 보임

- 분석과정보다는 결과에 집중하기! >> 분석 과정은 간단히 나타내기 (한장정도로), 분석 결과는 다양한 시각화를 활용하여 핵심만 작성하기
- 각 분석 별로 다양한 점수 산출 및 지표를 활용하여 타당성을 확보함 -> 시간만 많으면 먼저 다양한 연구를 읽어보고 어떤 분석을 진행했는지 이에 대한 공부를 먼저 수행하는 것이 좋을 듯 (한 3-4일 정도는 어떤 논문이 있는지, 논문에서 어떤 식으로 분석하고 어떤 지표를 활용하는지 선행연구 확인 필수 -> 이에 대한 근거는 반드시 출처 표시)
- 분석 결과는 되도록 시각화로 눈에 보이는 결과물로 나타내는 것이 좋음, 최종 결과만 굵은 글씨로 간단하게 결과 작성
- 마지막에 기대효과와 방향을 제시하면서 어떤식으로 해당 분석을 이용할 수 있는지를 보여주는 것이 확실히 활용성 방면에서 기승전결 구조를 갖추고 있어보임, 추가적으로 이에 대한 서비스 프로토타입을 보여주는 것도 좋아보였음
- 대상작 답게 굉장히 구조가 잘짜여져 있고, 누가봐도 이 팀은 분석을 정말 많이했다가 보임! 분석을 총 3가지로 나누어 각각분석하면서 파트분배도 각 공급, 수요, 상권분석 3가지로 명확하게 나눌 수 있을 듯
- 분석을 맡으면 해당 분석을 먼저 왜 하는지, 이 분석을 통해 나타난 결과를 다음 분석과 어떤식으로 연관을 지을지 전체적인 파이프라인으로 먼저 정리를 하고 진행하는 것이 좋을 것 같음
