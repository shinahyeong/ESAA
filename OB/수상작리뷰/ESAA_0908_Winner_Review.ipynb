2025 통계데이터 활용대회 최우수상 보고서

링크: https://data.kostat.go.kr/sbchome/bbs/boardDetail.do

- 주저: 빈집 발생의 다차원적 진단과 관리 전략: ESG를 중심으로
- 전체적인 progress
  - 분석 배경 및 목표 -> 데이터 분석 -> 예측 모델링 지수 -> 기대효과 및 활용방안

(1) 분석 배경 및 목표
- 사회 문제를 분석 -> 어느 지역의 빈집 문제가 심화되었는지, 지수를 왜 산출해야 하는지를 나타냄(회귀분석)
  - 분석이나 지수를 산출하기 전 "왜 이 지수/분석을 수행하는지?"에 대한 근거 마련 필요성 확인
- 목표를 기존 연구방식과 본 연구에서 진행할 진행방식을 비교하도록 한 슬라이드에 나타냄 -> 기대효과도 같이 언급
- 전체적으로 대상과 동일하게 로드맵 작성을 진행함 -> 한 페이지에 내가 이 분석을 왜 수행하게 되었는지, 해당 분석을 어떤식으로 진행할지, 분석 목표는 무엇인지, 총체적으로 이 분석을 통해 달성하고자 하는 바가 무엇인지를 한눈에 확인 가능
  - 전체적으로 이러한 로드맵을 작성함으로써 분석이나 보고서를 작성할 때 흐름을 구성하기 좋을 것 같다고 생각함

(2) 데이터 분석
- 대상과 마찬가지로 분석 과정의 파이프라인 (데이터 수집 -> 데이터 전처리 ->  데이터 분석 -> 모델링)을 한 슬라이드 형태로 담음
  - 본격 분석을 보여주기 전 내가 이 분석의 전체적인 요약본을 보여줌으로써 대략적으로 분석의 흐름을 유추할 수 있도록 함 (완성도가 높아보임)
  - 단 대상에 비해 색이 너무 화려하고 글씨 및 그림이 많음 -> 이런 부분은 감점 요소가 될 것 같기도?
  - 다양한 파생변수를 생성 후 EDA 수행 -> 비율의 차이가 존재함을 확인하며 가설이 성립함을 확인 (검정을 다양하게 활용하여 한가지 검정에서만 해당 결과가 나타난 것이 아닌, 다양한 검정에서 나타남을 보여줌 -> 타당성 확보)
  - 이런 사회문제에서 많이 보이는게 공간 자기 상관분석과 같은 공간분석이 많이 보임 -> 지도에 해당 구역을 표시하고 이런게 타당성 확보에 좋음 (이런 분석법을 추가저그로 공부해야 할 필요성을 느낌)
(3) 빈집 예측 모델링/지수
- 연구 목적을 설정한 후, 이에 적합한 모델을 생각하고, 대부분은 모델 성능 평가를 통해 성능이 높게 나오는 모델을 체택하는 방식을 이용하는 듯
- 해석 가능성을 강화하기 위한 SHAP분석을 수행
- 뭔가 다양한 분석을 수행하고 이 분석에 대한 정의나 아니면 분석에 대한 기초 개념도 많이 넣음
  - 약간 다소 줄 글이 많아보임
(4) 기대효과 및 활용 방안
- 크게 3가지의 기대효과를 들어 이에 대한 근거를 나타냄 -> 시각화부분이 직접 한 시각화를 들고와서 인상 깊어 보임
- 만든 지수를 어떤식으로 이용할지 구체적으로 나타내고, 이를 최종적으로 서비스 웹앱에 넣었을 때 어떻게 활용되는지 프로토 타입을 보여줌
  - 왠만해선 다 실현 가능성이 있어야 높은 점수를 받는듯

총평

[장점]
- 저번주 대상을 분석한 것과 동일하게 분석을 굉장히 체계적이고 많이 수행함
  - "왜?"라는 질문에 대해 모든 내용이 다 들어 있었음
- 기승전결이 명확하게 보여져서 완성도가 높아보임
- 다양한 로드맵, 파이프라인 등을 보여주어 이들이 어떤 분석을 수행할지, 수행했는지를 한눈에 보여줌 -> 해당 2페이지만 읽어도 대부분의 내용을 이해할 수 있음(마찬가지로 완성도가 높아보이는 요소로 작용)

[개선점]
- 너무 글이 많아보임 -> 발표에선 어떤 식으로 진행하였는지는 모르나, 해당 부분만 보면 집중력이 떨어짐 (대상이 안된 이유일 수도)
- 색이 너무 많음 -> 깔끔함이 다소 떨어짐 (색을 단일화 하는 것이 가독성이나 청중이 보기에는 더욱 좋아보일듯)
