제 11회 2022 빅데이터 공모전

주제: 탄소중립 선도를 위한 서울시 종합에너지스테이션 우선 입지 선정

흐름
1. 주제 선정 배경
- 탄소 중립 추구 -> 주유소 폐업이 잇달음
- 이에 대한 해결책으로 종합에너지 스테이션(TES)를 등장시켜, 탄소중립정책에 대응가능할 것
- 서울시 종합에너지스테이션 확대 계획을 펼침에 따라 우선 입지 분석이 필요하고, 이에대한 명확한 가이드라인 필요성이 대두됨

2. 목적 및 필요성
- 서울시 행정동별 전력 수급 및 전기차 충전소 수요를 고려하여 종합에너지스테이션 우선입지 행정동 및 장소 제공을 하고자 함

3. 유사연구사례 및 개선점
- 최적의 종합에너지스테이션 입지 분석을 위해 기존의 전기차 충전소, 수소차 충전소에 관한 선행 연구를 참고하고자 함 (총 4편 정도를 참고함)
- 선행연구 분석을 통해 이에 대한 개선점을 확인하고, 이를 3가지로 나타냄
  - 주요소 감소 추세방향
  - 인근 편의시설을 고려하여 이용자의 편의성 극대 방안 마련

4. 순서도
(1) 데이터 수집 및 전처리

(2) 클러스터링

(3) 입지 기준 선정

(4) 우선 입지 선정


2. 데이터 수집 및 전처리
- 행정동데이터: 21년 데이터 기준으로 행정동을 나타냄 -> 반영이 되지 않는 경우가 있어 면적을 기준으로 이에 대한 추가 행정동을 반영함
- 인구데이터: 기존 데이터셋 + 거주 데이터+ 거주 대비 비율
- 주유소 : 주유소 개수 및 전기 사용량 데이터를 가져옴
- 접근성(파생변수로 만든 데이터): 충전 시간이 비교적 오래 걸리는 경우 인근 편의시설이 존재하면 방문률이 높을 것이라 생각 -> 카카오 API를 토해 주유소 또는 충전소를 기준으로 500m 반경에 있는 음식점, 카페, 문화시설등의 개수를 추출함
- 전기차 데이터: 충전소 개수, 전기차 대수 등 뿐만 아니라 전기차 보급률, 처한대당 충전소 개수, 상권발달 등을 고려함 (파생변수화 )

3. EDA
- 주거 시각화 수행 -> 특정 행정동의 값들이 상대적으로 두드러지는 것을 확인 (이들을 중심으로 클러스터링, 입지분석 수행 )

4 클러스터링
- 주유소 생존 전략이므로, 이에 맞춰 주유소 개수가 0인 행정동은 제거
- 이후 267개의 행정동 대상으로 입지분석을 진행
- 다시한번 종합에너지스테이션의 특징을 다시 확인하여 변수들의 수요를 확인하기 위해 생활인구, 전기 사용량으로 클러스터링 수행
- 총 4가지로 클러스터링 수행
  - K-Means
  - K-medoids
  - HC
  - GMM
  
  -> 각 클러스터링 방법별 Elbow 기법과 Silhouette 계수 파악을 통해 최적의 클러스터 개수 도출 후 클러스터링 진행

  - 각 클러스터링을 종합적으로 고려하기 위해 Hard Voting 방법 수행 (2번이상 나온 행정동들을 대상으로 핵심 타겟 선정 )
    - 15개가 추출 : 유의미한 차이 발견

5. 입지 기준 도출
(1) 급속 전기차 충전기 설치
- 입지 기준을 설정하고자 많은 변수 중 Y= 급속일 때 이를 가장 잘 설명하는 변수 확인 -> 급속 충전기 개수가 0인 행을 제거함
- 다중공선성 제고: 너무 비슷한 변수 제거
- Best Selection을 하기 위해 이들의 변수 개수에 과민하게 반응하는 BIC 변수를 활용하여 변수 개수를 줄임
- 잠재요인 파악을 위해 요인 분석을 수행 -> 2가지 새로운 잠재요인 분석
- 최종 변수 : 급속 충전기 잠재요인(생활활성화), 충전환경, 상궈ㄴ발달, 급속충전기 결핍률
- PCA를 이용하여 평가지표 개를 개발함
  - 각 변수별로 가중치를 주어 차등적으로 부여한 후, 로딩값을 따와 최종적인 입지 지수를 적용함

(2) 입지분석 - 주유소 선정
- 입지 지수를 통해
  - 각 주요소에서 가장 가까운 충전소까지의 거리
  - 접근성 반영
