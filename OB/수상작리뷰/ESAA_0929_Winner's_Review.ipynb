## TEXOM 빅데이터 분석공모전 수상작
**우수상: 전기자동차 레드오션(red ocean) 시장인 중국에서 살아남는법**
- 여기선 텍스트 분석을 TEXOM, TEXOM 차이나를 이용하여 수행함
- 분석 방법
  - 약 1년 가량의 수집기간을 기준으로, 한국과 중국의 전기 자동차 관련 텍스트에서 어떤 텍스트가 많이 검색되어지고 있는지를 확인함
  - 한국: 네이버, 다음, 구글, 유튜브를 통해 '전기자동차' + window size 5 한국이라는 키위드로 분석을 수행
  - 중국: Baidu, CNKI, Sinaweibo, ⼈民网를 통해 '电动汽车(중국어로 전기 자동차라는 의미)'라는 키워드로 분석을 수행
- 둘에서 빈도수 10개의 단어를 비교하니, 두 나라 모두 자국 브랜드가 순위에 존재하는 것을 확인 가능
  - 한국: 기아, 르노코리아 자동차
  - 중국: 샤오미, 비야디
  
  -> 대부분 자국 전기자동차를 선호하는 경향을 확인할 수 있음
  - 또한 단어 빈도 10개에서 전기자동차를 수출입하는 국가의 언급도 많다는 사실을 확인 가능함
  - 중국은 이미 시장이 크기 때문에, 중국에서는 수출입되는 국가의 언급이 더욱 많은 반면, 아직 전기자동차의 수요가 작고, 인프라가 부족한 한국에서는 배터리, 충전과 같은 부속품이나 자동차의 존속에 대한 단어들의 언급도 잦은 것을 확인 가능
- 이를 워드클라우드, CONCOR 분석을 수행하여 각 단어의 언급 빈도수를 시각적으로 확인하고, 이를 CONCOR을 이용해 그룹화를 수행한 뒤, 이에 대한 자료를 찾아 근거를 설명한 것으로 나타남
  - 중국의 워드클라우드, CON
  - 텍스트 분석공모전에서는 텍스트를 수집하는 것도 중요하지만 이에 대한 근거를 찾는 과정이 더더욱 중요하는 것을 확인 가능 -> 분석 모델을 만드는 모델링 작업이 중요한 타 공모전에 비해 분석은 간결하나, 자료조사에 집중한 것을 확인 가능했음

- 결론 부분도 이 2가지 분석을 중심으로 자료조사를 통해 탄탄한 보고서를 구성한 것을 확인 가능

- 느낀점
  - 텍스트 분석에 대해 공부하면서 텍스트 분석이라는게 감정분석과 비슷하게 우리가 남긴 영화평 등의 글자로 표현된 감정을 직접 분석할 수 있다는 점에서 굉장히 흥미로웠는데, 공모전 수상작을 보면서 텍스트 분석에서의 핵심은 모델링이나 분석을 잘하는 것도 중요하지만 무엇보다 이걸 어떤식으로 풀어나갈건지 풀어나가는 과정이 중요한 분석기법이 아닐까?라는 생각을 하게됨
  - 분석방법은 간단하게 워드클라우드, CONCOR기법이었지만, 이 두가지 분석을 통해서도 다양한 인사이트를 얻을 수 있다는 점이 굉장히 배울점이라고 생각함.
  
    -> 굉장한 자료조사의 노력이 보여지는 보고서였음. 추가적으로 해당 분야에 대한 사전지식이 필수적임을 다시한번 깨닫게 됨

  - 보고서를 제출해야 하는 공모전의 경우, 화려한 모델링, 분석기법이 중요한것이 아닌, 이 기법을 왜 사용했고, 이 분석 방법을 통해 어떠한 인사이트를 얻을 수 있었는지 가 굉장히 중요하다는 것을 다시한번 알게됨
    - 공모전에 나갈 때에도 이전 인턴때 느낀 것과 동일하게 분석의 전처리 과정이나 내가 원하는 결과를 얻기 위해 핵심적인 모델링 프로세스를 구성하는 것도 중요하지만, 무엇보다 내가 이 분야, 이 공모전에서 선정한 주제를 얼마나 잘 알고 있는지가 중요하다고 생각함

      -> 지식의 싸움이라는 말이 적절한 표현인 것 같음
